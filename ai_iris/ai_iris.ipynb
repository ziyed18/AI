```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#=======================================
#1. Importation des librairies
#=======================================

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

#=======================================
#2.chargement du datasset Iris
#=======================================

iris=load_iris()
x=iris.data
y=iris.target
df=pd.DataFrame(x, columns=iris.feature_names)
df['species']=y
print("Aperçu du dataset Iris")
display(df.head())

#=======================================
#3.Séparation donnees train/test
#=======================================

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("Taille du train 1", x_train.shape)
print("Taille du test", x_test.shape)

#=======================================
#4.Entraimment du modele Random Forest
#=======================================

model=RandomForestClassifier(
    n_estimators=100,
    random_state=42)
model.fit(x_train,y_train)

#=======================================
#5.Prédiction et évluation (twa93)
#=======================================

y_pred=model.predict(x_test)
print("Accuracy :",accuracy_score(y_test,y_pred))
print("rapport de classification")
print(classification_report(y_test,y_pred,target_names=iris.target_names))

#=======================================
#6.Martice dde confusion
#=======================================

cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Matrice de confusion')
plt.xlabel('Prediction')
plt.ylabel('Réelles')
plt.show()

#=======================================
#7.Imortance deds variables
#=======================================
```
